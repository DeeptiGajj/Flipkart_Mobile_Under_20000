```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

```

```python
Product_Name = []
Price = []
ratings = []
```

```python
for j in range(2,12):
    url= "https://www.flipkart.com/search?q=mobile+under+20000rs&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_3_9_na_na_ps&otracker1=AS_Query_OrganicAutoSuggest_3_9_na_na_ps&as-pos=3&as-type=RECENT&suggestionId=mobile+under+20000rs&requestId=0db8a865-8131-45d3-b701-f96247453ce8&as-searchtext=mobile+un&page="+str(j)
  
    response = requests.get(url)
    soup= BeautifulSoup(response.content, "html.parser")
    

    boxdata = soup.find('div', class_ = "_1YokD2 _3Mn1Gg")
    
    name = boxdata.find_all('div',class_="_4rR01T")
 
    for j in name:
        j = j.get_text().replace('\n',"")
        j = j.strip(" ")
        Product_Name.append(j)
        
    scraped_Price = boxdata.find_all('div', class_="_30jeq3 _1_WHN1")
   
    for j in scraped_Price:
        j =  j.get_text().replace('\n',"")
        j =  j.strip(" ")
        Price.append( j)
         
    
    scraped_ratings = boxdata.find_all('div',  class_= "_3LWZlK")
   
    for j in scraped_ratings:
        j =  j.get_text().replace('\n',"")
        j =  j.strip(" ")
        ratings.append(j)
        ratings  
   
    
```

```python
data = pd.DataFrame()
data['Mobile_names'] = Product_Name
data['Price']= Price
data ['Ratings'] = ratings

data
```
